```python
from typing import (
    List,
)

class Solution:
    """
    @param a: an integer rotated sorted array
    @param target: an integer to be searched
    @return: an integer
    """
    def search(self, a: List[int], target: int) -> int:
        # One Binary Search
        if not a:
            return -1
        start, end = 0, len(a) - 1
        while start + 1 < end:
            mid = (start + end) // 2
            if a[mid] > a[end]:
                if a[start] <= target <= a[mid]:
                    end = mid
                else:
                    start = mid
            else:
                if a[mid] <= target <= a[end]:
                    start = mid
                else:
                    end = mid
        if a[start] == target:
            return start
        if a[end] == target:
            return end
        return -1

        """
        if not a:
            return -1
        # Find the rotation point
        start, end = 0, len(a) - 1
        while start + 1 < end:
            mid = (start + end) // 2
            if a[mid] > a[end]:
                start = mid
            else:
                end = mid
        if a[start] <= a[end]:
            rotate_point = start
        else:
            rotate_point = end

        # Find the target index
        if target > a[-1]:
            start, end = 0, rotate_point - 1
        else:
            start, end = rotate_point, len(a) - 1
        while start + 1 < end:
            mid = (start + end) // 2
            if a[mid] > target:
                end = mid - 1
            elif a[mid] == target:
                end = mid
            else:
                start = mid + 1
        if a[start] == target:
            return start
        if a[end] == target:
            return end
        return -1
        """
```
