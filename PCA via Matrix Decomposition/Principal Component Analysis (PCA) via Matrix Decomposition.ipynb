# Population PCA
Population PCA is about trying to transform the random variables $X_1, \dots, X_p$ linearly so that the new random variables $Y_1, \dots, Y_p$ are 

- uncorrelated such that Cov($Y_i$, $Y_j$) = 0 for i $\neq$ j
- variances are as large as possible such that Var($Y_i$) is maximized  for i = $1, \dots, p$

Then, $Y_1, \dots, Y_p$ are called population principal components which can be written as:
   
$$Y_1 = a_{11}X_1 + \dots + a_{1p}X_p$$
$$\vdots$$
$$Y_p = a_{p1}X_1 + \dots + a_{pp}X_p$$
    
This is equivalent to the matrix expression that $Y=AX$

To find $Y_1, \dots, Y_p$, we find them one by one:
- 1st principal component: $$Y_1 = a_1^Tx, \quad a_1 = argmax\{Var(a^Tx): ||a||_2=1\}$$
- 2nd principal component: $$Y_2 = a_2^Tx, \quad a_2 = argmax\{Var(a^Tx): ||a||_2=1, \quad Cov(a^Tx, a_1^Tx)=0\}$$
- 3rd principal component: $$Y_3 = a_3^Tx, \quad a_3 = argmax\{Var(a^Tx): ||a||_2=1, \quad Cov(a^Tx, a_i^Tx)=0  \quad i=1,\dots,k-1\}$$
- ...

### Theorem:

Let $\Sigma = Cov(X) \in R^{p\times p}$ and apply the eigenvalue decomposition (EVD) such that $\Sigma = Q\Lambda Q^T$ where $Q$ = [$q_1$, \dots, $q_p$] $\in R^{p\times p}$ is orthogonal and $\Lambda = diag(\lambda_1, \dots, \lambda_p)$ with $\lambda_1 \ge \lambda_2 \ge\dots\ge\lambda_p \ge 0$

Then,<br>
$\qquad Y_k = q_k^Tx, k = 1, \dots, p$ are the population principal components of X.<br>
Furthermore,<br>
$\qquad Var(Y_k) = q_k^T\Sigma q_k = \lambda_k$<br>
and<br>
$\qquad Cov(Y_i, Y_j) = q_i^T\Sigma q_j = 0$ for $i\ne j$

Essentially, we find that the population PCA on X is equivalent to the EVD of the covariance matrix $\Sigma$ of X

### Properties:
- Total variance is constant:
$$tr(\Sigma) = tr(\Lambda)$$
$$\sigma_{11} + \dots + \sigma_{pp} = \lambda_1 +\dots + \lambda_p$$
$$Var(X_1) + \dots + Var(X_p) = Var(Y_1) + \dots + Var(Y_p)$$
- The proportion of total variance attributable to the kth principal component $Y_k$ is given by:
$$\frac{\lambda_k}{\lambda_1 + \dots + \lambda_k} * 100\%$$



# Sample PCA
In reality, we don't have random variables. Instead, we have data X such that
$$X = [x_1^T, ..., x_n^T]\in R^{n\times p}$$
where $x_i$ is the ith observation.

Analogous to Population PCA in that we want to find $a\in R^p$ such that 
- $a^Tx_1, ..., a^Tx_n$ has the maximum variance
- and that $a^Tx_1, ..., a^Tx_n$ are uncorrelated with $b^Tx_1, ..., b^Tx_n$ for other sample PC.

Thus,
- 1st principal component: <br>
$a_1^T \in R^p$ such that the sample variance of $a_1^Tx_1, ..., a_1^Tx_n$ is maximum among all possible linear combinations, i.e.,
$$a_1 = argmax\{a^TSa: ||a||_2=1\}$$ where S is the sample covariance matrix
- 2nd principal component: <br>
$a_2^T \in R^p$ such that the sample variance of $a_2^Tx_1, ..., a_2^Tx_n$ is maximized and subjected to $a_1^Tx_1, ..., a_1^Tx_n$ and $a_2^Tx_1, ..., a_2^Tx_n$ are uncorrelated.
$$a_2 = argmax\{a^TSa: ||a||_2=1, a^TSa_1 = 0\}$$
- ...
- kth principal component: <br>
$$a_k = argmax\{a^TSa: ||a||_2=1, a^TSa_i = 0 \quad i = 1,..., k-1\}$$

### Properties:
- Similar to population PC, total sample variance is:
$$S_{11} + \dots + S_{pp} = tr(S)$$
- The proportion of total sample variance due to kth principal component is:
$$\frac{\lambda_k}{\lambda_1 + \dots + \lambda_k} * 100\%$$
- We can also do correlation PCA. Apply sample PCA using standardized data matrix Z ("EVD on Cov(Z)") is equivalent to doing EVD on the sample correlation matrix P of X.


### Theorem:

Let $S \in R^{p\times p}$ be the sample covariance matrix and apply the eigenvalue decomposition (EVD) such that $S = Q\Lambda Q^T$ where $Q$ = [$q_1$, ..., $q_p$] $\in R^{p\times p}$ is orthogonal and $\Lambda = diag(\lambda_1, ..., \lambda_p)$ with $\lambda_1 \ge \lambda_2 \ge...\ge\lambda_p \ge 0$

Then, the kth PC of $x_1, ..., x_n$ is given by $q_k$ for $k = 1,..., p$

Furthermore,the sample variance of<br>
$\qquad q_k^Tx_1, ..., q_k^Tx_n$ is $\lambda_k$ for $k = 1, ..., p$ <br>
and the sample covariance of<br>
$\qquad q_i^Tx_1, ..., q_i^Tx_n$ and $q_j^Tx_1, ..., q_j^Tx_n$ is zero for $i\ne j$

Essentially, we find that the population PCA on X is equivalent to the EVD of the its sample covariance matrix $S$

# PCA Using Singular Value Decomposition (SVD)
### Theorem:
Given the data X, we can compute SVD of the mean centered data $X_c$. This is equivalent to the EVD of the covariance of the data X.

\begin{align}
S &= \frac{1}{n-1}(X-1x^T)(X-1x^T) \\
&= \frac{1}{n-1}X_c^TX_c\\
&= \frac{1}{n-1}V\Sigma^T\Sigma V^T \\
&= V\left(\begin{array}{cc} 
\frac{\sigma_1^2}{n-1} & 0 & ... & 0\\
0 & \frac{\sigma_2^2}{n-1} & ... & 0\\
\vdots & \vdots & \ddots & 0\\
0 & 0 & \cdots & \frac{\sigma_p^2}{n-1}
\end{array}\right)V^T
\end{align}
We can denote $Q = V$, $\Lambda=\frac{1}{n-1}\Sigma^T\Sigma$ such that $S=Q\Lambda Q^T$

### Remark:
- $Q = V$: the sample principal components are the right singular vectors of $X_c$
- $\lambda_k = \frac{\sigma_k^2}{n-1}$: the sample variance of $v_k^Tx_1, ..., v_k^Tx_n$
- The projection of ith data point $x_i$ onto jth principal component $v_j$ is $\sigma_j u_{i_j}$

### Two Reasons why SVD of X_c is preferrable to EVD of S:
- To get S, we need to compute $Y^TY$ which is unstable. Computing SVD of X_c is numerically more robust.
- We can get all possible scatter plots for all j,k at once just from the score matrix $T = U\Sigma$

Therefore, if we want PCA of the data $X$, we usually compute SVD of the mean centered data $X_c$ where $X_c = X - 1x^T$ 

# Correlation PCA
EVD applied to the correlation matrix P is called correlation PCA. It is equivalent to first standardizing and mean centering on the random variables
$$X_1, \dots, X_p \to Z_1, \dots, Z_p \quad where\quad Z_i = \frac{X_i-\mu_i}{\sqrt{\sigma_{ii}}}$$
because
$$Z = V^{-1/2}(X-\mu)$$
$$Cov(Z) = V^{-1/2}\Sigma_XV^{-1/2}=P_X=Corr(X)$$

# Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =  pd.read_csv(r"raw.txt")
```

```python
data
```

```python
data.shape
```

```python
processed = (data - data.mean(axis=0)) / data.std(axis=0)
```

```python
processed
```

```python
u,s,vt = np.linalg.svd(processed)
print('u shape:', u.shape)
print('s shape:', s.shape)
print('vt shape:', vt.shape)
```

```python
lambdas = [sigma**2/(len(processed)-1) for sigma in s]
lambdas
```

```python
plt.plot(lambdas,'.-')
plt.xlabel('Components')
plt.ylabel('Eigenvalues')
plt.title('Scree Plot')
```

## Plot of the PC1 and PC2

```python
sigma_1, sigma_2 = s[0], s[1]
u_1, u_2 = u[:,0], u[:,1]
x, y = sigma_1*u_1, sigma_2*u_2

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.figure
plt.xlabel('$v_1$')
plt.ylabel('$v_2$')

countries = data.index
for i, country in enumerate(countries):
    plt.annotate(country, (x[i], y[i]))
```

- In this graph, x and y axes correspond to $v_1$ and $v_2$ respectively and the points correspond to the countries
- We observe that Singapore and Hong Kong are two outliers as they are apart from the other countries

## Plot of the Variables

```python
sigma_1, sigma_2 = s[0], s[1]
v_1, v_2 = vt[0], vt[1]
x, y = sigma_1*v_1, sigma_2*v_2

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.figure
plt.xlabel('$u_1$')
plt.ylabel('$u_2$')

demographic_variables = data.columns
for i, country in enumerate(demographic_variables):
    plt.annotate(country, (x[i], y[i]))
```

- In this graph, x and y axes correspond to $u_1$ and $u_2$ respectively and the points correspond to the demographic variables.

## Biplot

```python
# biplot (no good theory behind this)
sigma_1, sigma_2 = s[0], s[1]
u_1, u_2 = u[:,0], u[:,1]
x, y = sigma_1*u_1, sigma_2*u_2

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.figure
plt.xlabel('v1 & u1')
plt.ylabel('v2 & u2')
plt.xlim((-7,7))
plt.ylim((-7,5))

countries = data.index
for i, country in enumerate(countries):
    plt.annotate(country, (x[i], y[i]))
    
v_1, v_2 = vt[0], vt[1]
for x,y,var in zip(sigma_1*v_1, sigma_2*v_2, demographic_variables):
    plt.scatter(x, y, color='r')
    plt.arrow(0, 0, x, y,color = 'r',alpha = 0.5)
    plt.text(x*1.05, y*1.05, var, color = 'b', ha = 'center', va = 'center')
```

- By overlaying the previous two graphs, we can identify the demographic variables near the two outlier countries.
- Singapore and Hong Kong are two outliers because they have extreme values in "Population/sq.km" and "Population/1000HectarAgri"

## Remove the Two Outlier Countries and Redo the Previous Analysis

```python
processed_dropped = processed[~processed.index.isin(['Hong Kong', 'Singapore'])]
```

```python
u,s,vt = np.linalg.svd(processed_dropped)
print('u shape:', u.shape)
print('s shape:', s.shape)
print('vt shape:', vt.shape)
```

```python
sigma_1, sigma_2 = s[0], s[1]
u_1, u_2 = u[:,0], u[:,1]
x, y = sigma_1*u_1, sigma_2*u_2

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.figure
plt.xlabel('$v_1$')
plt.ylabel('$v_2$')

countries = processed_dropped.index
for i, country in enumerate(countries):
    plt.annotate(country, (x[i], y[i]))
```

- By removing the two outlier countries, we can see features that were earlier obscured by the outliers.
- For example, we can find that Finland and Austria are two European countries most alike Japan.

```python
mask = countries == 'Japan'
japan = (x[mask][0], y[mask][0])
distance_to_japan = [(country, ((x[i]-japan[0])**2 + (y[i]-japan[1])**2)**(1/2)) for i, country in enumerate(countries)]
distance_to_japan.sort(key = lambda x: x[1])
```

```python
pd.DataFrame(distance_to_japan, columns=['Country', 'Distance to Japan'])
```
